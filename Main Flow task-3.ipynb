```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv('householdtask3.csv')
```

```python
df.head()
```

```python
plt.scatter(df['year'],df['own'])
plt.title('Scatter Plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.plot(df['year'])
plt.plot(df['own'])
plt.title('Line Plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.bar(df['year'],df['own'])
plt.title('Bar Plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.hist(df['income'])
plt.title('Histogram')
plt.show()
```
